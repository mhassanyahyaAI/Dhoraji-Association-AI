```python
student = pd.DataFrame ({'Name':['Ahsan','Umer',"Ali", "Taha","Bilal"],
                         "Age":np.random.randint(13,25,size=[5]),
                         "Percentage":np.random.randint(50,90,size=[5])})
np.random.seed(5)
```

```python
student
```

```python
def name(age):
    if age >18:
        return 'Adult'
    else:
        return 'Non-Adult'
```

```python
student['Status'] = student ['Age'].map(name)
```

```python
student
```

```python
def name(percentage):
    if percentage >70:
        return 'Pass'
    else:
        return 'Fail'
```

```python
student['Result'] = student ['Percentage'].map(name)
```

```python
student
```

```python
student[student.Status == 'Adult']
```

```python
student[student.Result == "Pass"  "Adult"]
```
